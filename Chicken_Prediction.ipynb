```python
from tensorflow.keras.models import load_model

model = load_model('/content/drive/MyDrive/chicken_dataset/chicken_breed_classifier_mobilenetv2.h5')
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Change this to the name of your uploaded image
img_path = list(uploaded.keys())[0]

# Load and preprocess
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0  # normalize

# Predict
prediction = model.predict(img_array)
predicted_class = np.argmax(prediction[0])

# Display image
plt.imshow(img)
plt.axis('off')
plt.title(f"Predicted class index: {predicted_class}")
plt.show()
class_names = ['Bielefeld', 'Blackorpington', 'Brahma', 'Buckeye', 'Fayoumi',
               'Leghorn', 'New Hampshire', 'Plymouth Rock', 'Sussex', 'Turken']
print("Predicted class name:", class_names[predicted_class])
```
